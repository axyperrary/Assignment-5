# Pyber Analysis
- The data shows that as popluation become more urban the average fare for a ride decreases. 
- The data also shows that as the number of total rides increase in urban areas so does the total fares and total       drivers.
- In conclusion, the data shows that the more people there are in a city the more rides and thus fares there will be,   however; there are some larege cities with high rides rates but low average fares.

```python
%matplotlib inline

# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
data_df = pd.merge(ride_data_df, city_data_df)

# Display the data table for preview
data_df.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
rural_df = data_df.loc[data_df["type"] == "Rural", :]

rural_cities = rural_df.groupby("city")

x_axis = rural_cities["fare"].mean()

driver_count = rural_cities["driver_count"].mean()

total_rides_city = rural_cities["ride_id"].unique()
for i in range(len(total_rides_city)):
    total_rides_city[i] = len(total_rides_city[i])
y_axis = total_rides_city

urban_df = data_df.loc[data_df["type"] == "Urban", :]
urban_cities = urban_df.groupby("city")
u_x_axis = urban_cities["fare"].mean()
u_driver_count = urban_cities["driver_count"].mean()
total_rides_city = urban_cities["ride_id"].unique()
for i in range(len(total_rides_city)):
    total_rides_city[i] = len(total_rides_city[i])
u_y_axis = total_rides_city

suburban_df = data_df.loc[data_df["type"] == "Suburban", :]
suburban_cities = suburban_df.groupby("city")
su_x_axis = suburban_cities["fare"].mean()
su_driver_count = suburban_cities["driver_count"].mean()
total_rides_city = suburban_cities["ride_id"].unique()
for i in range(len(total_rides_city)):
    total_rides_city[i] = len(total_rides_city[i])
su_y_axis = total_rides_city

# Build the scatter plots for each city types
rural_plt = plt.scatter(y_axis, x_axis, s = driver_count * 10, facecolors = "gold", edgecolor="black", alpha = 0.7,
                        linewidth = 1.5, label = "Rural")

urban_plt = plt.scatter(u_y_axis, u_x_axis, s = u_driver_count * 10, facecolors = "lightcoral", edgecolor="black", 
                        alpha = 0.7, linewidth = 1.5, label = "Urban")

suburban_plt = plt.scatter(su_y_axis, su_x_axis, s = su_driver_count * 10, facecolors = "lightskyblue", 
                           edgecolor="black", alpha = 0.7, linewidth = 1.5, label = "Suburban")
 
# Incorporate the other graph properties
plt.grid()

plt.title("Pyber Ride Sharing Data (2016)")
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")

# Create a legend
plt.legend(handles = [urban_plt, suburban_plt, rural_plt], loc = "best", title = "City Types")

# Incorporate a text label regarding circle size
plt.gcf().text(0.92, 0.6, "Note: \nCircle size coorelates with driver count per city.", fontsize = 12)

# Save Figure
plt.savefig("./RideSharingScatter.png")
```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
urban_total_fares = urban_df["fare"].sum()
surban_total_fares = suburban_df["fare"].sum()
rural_total_fares = rural_df["fare"].sum()

# Build Pie Chart
labels = ["Urban", "Suburban", "Rural"]
colors = ["lightcoral", "lightskyblue", "gold"]
explode = (0.1, 0, 0)
plt.pie((urban_total_fares, surban_total_fares, rural_total_fares), explode = explode, labels = labels, 
        colors = colors, autopct = "%1.1f%%", shadow = True, startangle = 290)
plt.title("% of Total Fares by City Type")

# Save Figure
plt.savefig("./RidesFaresPie.png")
```

```python
# Show Figure
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
types = data_df["type"].value_counts()
total = len(types)
types_percent = (types / total)
#types_percent
#total

# Build Pie Chart
labels = ["Urban", "Suburban", "Rural"]
colors = ["lightcoral", "lightskyblue", "gold"]
explode = (0.1, 0, 0)
plt.pie(types_percent, explode = explode, labels = labels, colors = colors, autopct = "%1.1f%%", shadow = True, 
        startangle = 260)
plt.title("% of Total Rides by City Type")

# Save Figure
plt.savefig("./RidesCityTypePie.png")
```

```python
# Show Figure
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Driver Percents
#print(urban_df["driver_count"].value_counts())
urban_drivers = urban_df["driver_count"].mean()
surban_drivers = suburban_df["driver_count"].mean()
rural_drivers = rural_df["driver_count"].mean()

# Build Pie Chart
labels = ["Urban", "Suburban", "Rural"]
colors = ["lightcoral", "lightskyblue", "gold"]
explode = (0.1, 0, 0)
plt.pie((urban_drivers, surban_drivers, rural_drivers), explode = explode, labels = labels, 
        colors = colors, autopct = "%1.1f%%", shadow = True, startangle = 200)
plt.title("% of Total Drivers by City Type")

# Save Figure
plt.savefig("./RidesDriversPie.png")
```

```python
# Show Figure
plt.show()
```
